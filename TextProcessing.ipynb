```python
from nltk import tokenize

corpus_dir = "Texts/"
```

```python
# Given a text file, returns a 2 dictionaries:
# Map of sentence_ID to sentences 
# Map of sentence_ID to set of all words in the sentence with the same ID
def process_text(file_name):
    text = open(file_name, encoding="utf-8").read()
    # Split into sentences
    sentences = tokenize.sent_tokenize(text)
#     print(sentences)
    s_id = 0
    file_sentences = {}
    file_wordsets = {}
    for s in sentences:
        words = set(s.strip().split())
        file_sentences[s_id] = s
        file_wordsets[s_id] = words
        s_id += 1
    return file_sentences, file_wordsets

process_text(corpus_dir + "DescartesMeditations.txt")
```

```python
# Getting and saving dicts for all texts
process_text(corpus_dir + "KewGardens.txt")
```
